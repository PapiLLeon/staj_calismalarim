```python
#======================
#GEREKLİ KÜTÜPHANELER
#======================
import datetime               #NOT:Göreve tarih ve saat eklemek için kullanılır.


# ======================
# AYARLAR
# ======================
DOSYA_ADI = "gorevler.txt"    #NOT:Dosya adını değişken yaptım böylece tek yerden değiştirilebilir.
gorevler = []                 #NOT:Görevlerin saklanacağı liste.


# ===========================
# ANA MENÜYE DÖNME FONKSİYONU
# ===========================
def devam_et():
    input("\nAna menüye dönmek için Enter basınız...")

# =====================================
# DOSYADAN GÖREVLERİ YÜKLEME FONKSİYONU
# =====================================
def gorevleri_yukle():
    try:
        with open(DOSYA_ADI, "r", encoding="utf-8") as dosya:
            for satir in dosya:
                gorev = satir.strip().split(",")
                if len(gorev) == 4:
                    gorevler.append({
                        "gorev": gorev[0],
                        "durum": gorev[1],
                        "oncelik": gorev[2],
                        "tarih": gorev[3]
                    })
    except FileNotFoundError:
        pass

# ==============================
# GÖREVLERİ DOSYAYA KAYDETME
# ==============================

def gorevleri_kaydet():
    with open(DOSYA_ADI, "w", encoding="utf-8") as dosya:
        for g in gorevler:
            satir = f"{g['gorev']},{g['durum']},{g['oncelik']},{g['tarih']}\n"
            dosya.write(satir)

#================================================
#GÖREVLERİ ÖNCELİKLERİNE GÖRE SIRALIYAN FONKSİYON
#================================================
#NOT:Yüksek öncelikli görevlerin yukarıda sıralanmasını sağlar.

def gorevleri_sirala():
    siralama = {
        "Yüksek": 1,
        "Orta": 2,
        "Düşük": 3
    }
    gorevler.sort(key=lambda g: siralama.get(g["oncelik"], 2)) #NOT:.sort() artan sırada sıralar.Yeni liste döndürmez 'none' döner. key:Karşılaştırmada kullanılacak değeri döndüren fonksiyonu alır. lambda:Tek satırlık anonim fonksiyon tanımlamak için kullanılır.


#==========================
#GÖREV LİSTELEME FONKSİYONU
#==========================

def gorevleri_listele():
    if not gorevler:
        print("Görev listesi boş.")
        return

    gorevleri_sirala()

    print("\n📋 GÖREVLER\n")

    for i, g in enumerate(gorevler, start=1):

        # NOT:Eğer görev tamamlandıysa farklı emoji gösteriyor.

        if g["durum"] == "Tamamlandı":
            durum_emoji = "✔"
        else:
            durum_emoji = "⏳"

        print(f"{i}. {g['gorev']}")
        print(f"   {durum_emoji} Durum   : {g['durum']}")
        print(f"   🔥 Öncelik : {g['oncelik']}")
        print(f"   🕒 Tarih   : {g['tarih']}")
        print("-"*35)

#=======================
#GÖREV EKLEME FONKSİYONU
#=======================

def gorev_ekle():

    gorev = input("Yeni görev giriniz: ")

    if gorev.strip() == "":
        print("Boş görev ekleyemezsiniz.")
        return

    oncelik = input("Öncelik (Yüksek / Orta / Düşük): ").capitalize()

    if oncelik not in ["Yüksek", "Orta", "Düşük"]:
        oncelik = "Orta"

    tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")

    gorevler.append({
        "gorev": gorev,
        "durum": "Yapılıyor",
        "oncelik": oncelik,
        "tarih": tarih
    })

    gorevleri_kaydet()

    print("✅ Görev eklendi.")

#==========================
#GÖREVLERİ SİLME FONKSİYONU
#==========================

def gorev_sil():

    gorevleri_listele()

    if not gorevler:
        return

    try:
        no = int(input("Silinecek görev numarası: "))

        if 1 <= no <= len(gorevler):

            silinen = gorevler.pop(no-1)

            gorevleri_kaydet()

            print(f"🗑 {silinen['gorev']} silindi.")

        else:
            print("Geçersiz numara.")

    except ValueError:
        print("Lütfen sayı giriniz.")

#==============================
#GÖREVLERİ DÜZENLEME FONKSİYONU
#==============================

def gorev_duzenle():

    gorevleri_listele()

    if not gorevler:
        return

    try:

        no = int(input("Düzenlenecek görev numarası: "))

        if 1 <= no <= len(gorevler):

            yeni = input("Yeni görev metni: ")

            if yeni.strip() == "":
                print("Boş görev girilemez.Lütfen görev metni giriniz.")
                return

            gorevler[no-1]["gorev"] = yeni

            gorevleri_kaydet()

            print("✏ Görev güncellendi.")

        else:
            print("Geçersiz düzenlenecek görev numarası.")

    except ValueError:
        print("Lütfen sayı giriniz.!")

#==============================
#GÖREVLERİ TAMAMLAMA FONKSİYONU
#==============================

def gorev_tamamla():

    gorevleri_listele()

    if not gorevler:
        return

    try:

        no = int(input("Tamamlanan görev numarası: "))

        if 1 <= no <= len(gorevler):

            gorevler[no-1]["durum"] = "Tamamlandı"

            gorevleri_kaydet()

            print("✔ Görev tamamlandı.")

        else:
            print("Geçersiz numara.")

    except ValueError:
        print("Lütfen sayı giriniz.")

# ========================
# ANA MENÜ FONKSİYONU
# ========================

def menu():

    print("\n  ╔══════════════════════╗")
    print("    GÖREV YÖNETİM SİSTEMİ")
    print("  ╚══════════════════════╝\n")

    print("1 📋 Görevleri Listele")
    print("2 ➕ Görev Ekle")
    print("3 ✏ Görev Düzenle")
    print("4 🗑 Görev Sil")
    print("5 ✔ Görev Tamamla")
    print("6 🚪 Çıkış\n")

    return input("Seçiminiz: ")

# Programın ana döngüsü
gorevleri_yukle()  # Program başladığında görevleri yükle

while True:

    secim = menu()

    if secim == "1":
        gorevleri_listele()
        devam_et()

    elif secim == "2":
        gorev_ekle()
        devam_et()

    elif secim == "3":
        gorev_duzenle()
        devam_et()

    elif secim == "4":
        gorev_sil()
        devam_et()

    elif secim == "5":
        gorev_tamamla()
        devam_et()

    elif secim == "6":
        print("Programdan çıkılıyor...")
        break

    else:
        print("Geçersiz seçim!")
        devam_et()
```
